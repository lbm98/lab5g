**Question:**
Enumerate and summarize the main phases (from power-on to RRC_CONNECTED) during the connection of a UE to the gNodeB?

**Answer:**

```
- The UE powers-on and is in RRC_IDLE state

- The UE sends the Preamble Message (Msg1) on the PRACH
    - The message contains a random sequence number

- The gNB responds with the Random Access Response (Msg2)
    - The message contains the TA command, the RAPID and the UL Grant for Msg3
    - The gNB also assigns a temporary identifier (RA-RNTI) to the UE

- The UE sends the RRC Setup Request Message (Msg3) on the PUSCH
    - The UE uses the UL Grant

- The gNB sends the RRC Setup Message (Msg4)
    - The message contains MAC data for Contention Resolution, like the UE's identity

- The UE receives the RRC Setup Message (Msg4) and transitions to the RRC_CONNECTED state
```

**Question:**
Describe the origin of OAI.

**Answer:**
The OpenAirInterface Software Alliance was
established in 2014, the OSA is a French non-profit organization (“Fonds De Dotation”), funded by corporate sponsors.

The OSA is the home of OpenAirInterface, an open software that gathers a community of developers from around the world, who work together to build wireless cellular Radio Access Network (RAN) and Core Network (CN) technologies.

**Sources:**
- https://openairinterface.org/about-us/

**Question:**
Discuss the licensing model for OAI. Is it really open source? Why?

**Answer:**
The OAI Public License V1.1 is a modified version of Apache V2.0 License, which is a permissive free software license that allows users to use the software for any purpose.
However, the modified patent clause that allows contributing parties to make patent license available to third parties under fair, reasonable and non-discriminatory (FRAND) terms for commercial exploitation.
Thus, contributors to the OAI source code are able to keep their patent rights, so the entire codebase is not open-source.


**Sources:**
- https://openairinterface.org/legal/oai-license-model/
- https://en.wikipedia.org/wiki/Apache_License
